```python
import seaborn as sns
fig=plt.figure(figsize=(50,20))
sns.barplot(data=df,
           x='Year',
           y='Number of Fires',
            palette='Blues_d')
plt.xlabel('Year',fontsize=40)
plt.ylabel('Number of Fires',fontsize=40)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.title(label="Number of Fires vs Year",fontsize=100)
```

```python
fig=plt.figure(figsize=(200,100))

sns.barplot(data=df1,
            x=df1.index,
            y=df1['Number of Fires'],
            palette='Blues_d')
plt.xlabel('State',fontsize=150)
plt.ylabel('Number of fire recorded',fontsize=150)
plt.xticks(fontsize=60)
plt.yticks(fontsize=75)
plt.title(label="Number of Fires vs Year",fontsize=100)
```
