```python
# Imputing Rating column
```

- Imputing Rating column first involves converting it into categorical variable and then use KNNImputer

```python
df['Rating'].unique()
```

```python
df
```

```python
# Extract the column to be imputed
rating_column = df['Rating'].copy()

# Convert categorical ratings to numeric labels for KNN imputation
rating_mapping = {'E': 0, 'M': 1, 'T': 2, 'E10+': 3, 'K-A': 4, 'AO': 5, 'EC': 6, 'RP': 7}
rating_column = rating_column.map(rating_mapping)
```

```python
# Create a copy of the DataFrame without the "Rating" column
imputation_df = df.drop(columns=['Rating']).copy()

# Initialize KNNImputer with one feature ("Rating")
imputer = KNNImputer(n_neighbors=5)
```

```python
# Perform KNN imputation on the "Rating" column
imputed_ratings = imputer.fit_transform(rating_column.values.reshape(-1, 1))
```

```python
reversed_mapping = {v: k for k, v in rating_mapping.items()}
imputed_ratings = [reversed_mapping[int(r)] for r in imputed_ratings]
```

- the code takes a mapping that maps original ratings to some codes, reverses it to create a mapping that maps codes back to the original ratings, and then uses this reversed mapping to transform a list of imputed rating codes into their corresponding original ratings.

```python
imputation_df['Rating'] = imputed_ratings
```

#### 8) Which year had the highest sales worldwide?

```python
# Group the dataset by 'Year_of_Release' and calculate the sum of global sales for each year
y_sales = df.groupby('Year_of_Release')['Global_Sales'].sum()

# Find the year with the highest global sales
y_sales_max = y_sales.idxmax()

# Print the year with the highest global sales
print("Year with the highest global sales:", y_sales_max)
```

#### 9) Which genre game has been released the most in a single year?

```python
# Group the dataset by 'Year_of_Release' and 'Genre' and calculate the count of games
year_release_genre_count = df.groupby(['Year_of_Release', 'Genre']).size()

# Find the genre with the highest count for each year
most_released_genre_game_year = year_release_genre_count.groupby('Year_of_Release').idxmax()
```

```python
print("The genre game has been released the most in a single year",most_released_genre_game_year)
```

#### 10) Which genre game has the highest sale price globally?

```python
# Group the dataset by 'Genre' and calculate the sum of global sales for each genre
genre_global_sales = df.groupby('Genre')['Global_Sales'].sum()

# Find the genre with the highest global sales
highest_global_sales = genre_global_sales.idxmax()
```

```python
print("Genre with highest global sales:", highest_global_sales)
```
