# Importing Libraries

```python
#Libraries

import pandas as pd

#to generate fake ids
import uuid

#random is required to generate a random set instead of a very organised one
import random

from faker import Faker
```

```python
# Number of users/rows to create
num_users=100 #more the number more random it is 
```

- Add feature list
      - ID
      - Gender
      - Subscriber
      - Name
      - Rating

# Creating a DataFrame

```python
# A list of five features

features=["id","gender","subscriber","name","rating"]

# Creating a dataFrame for these features
df=pd.DataFrame(columns=features)
df
```

# Generating fake data

- Generating imbalanced data
        - ideally data is equally represented but in reality data in imbalanced
       

### Generating fake id of users

```python
df['id']=[uuid.uuid4().hex for i in range(num_users)]
df
```

```python
# Checking if all IDs are unique

print(df['id'].nunique==num_users)
```

### Generating Gender of users

```python
genders=["Male","Female","Not specified"]

df['gender']=random.choices(genders,weights=(45,45,10),k=num_users)
df
```

### Generating subscription of users

```python
choice=[True,False]

df['subscriber']=random.choices(choice,k=num_users)
df
```

### Generating fake names of users

```python
faker=Faker(locale="EN_IN")

def name_generator(gender):
    
    if gender=='Male':
        return faker.name_male()
    elif gender=='Female':
        return faker.name_female()
    else:
        return faker.name

# Generating names for each user
df['name']=[name_generator(i) for i in df['gender']]

df
```

### Generating rating for the user

```python
rating=[1,2,3,4,5]

df['rating']=random.choices(rating,weights=(10,30,45,45,10),k=num_users)
df
```
